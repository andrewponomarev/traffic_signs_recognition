### 2.2. Model Architecture

````python
# Network params
dropout_conv = 0.9
dropout_fc = 0.9


kernel_size = (5,5)
conv_output = [6, 16]
fc_output = [120, 84, n_classes]

weights_stddev = 0.1
weights_mean = 0.0
````

````python
### Define your architecture here.

from tensorflow.contrib.layers import flatten
import tensorflow as tf

# Max pooling
def max_pool(x, pool_size=2, padding_value='VALID'):
    return tf.nn.max_pool(x, ksize=[1, pool_size, pool_size, 1], strides=[1, pool_size, pool_size, 1],
                          padding=padding_value)

# Convolutional 
def conv2d(x, weights, bias, strides = 1, padding_value='VALID'):
    return tf.nn.conv2d(x, weights, strides=[1, strides, strides, 1], padding = padding_value) + bias
    return tf.nn.relu(conv)

def weigths_variable_conv(kernel_size, in_channels, out_channels, weight_mean, weight_stddev):
    return tf.Variable(tf.truncated_normal(shape=(kernel_size[0], kernel_size[1], in_channels, out_channels), mean = weight_mean, stddev = weight_stddev))

def weights_variable_fc(in_channels, out_channels, weight_mean, weight_stddev):
    return tf.Variable(tf.truncated_normal(shape=(in_channels, out_channels), mean = weight_mean, stddev = weight_stddev))
    
def bias_variable(shape):
    return tf.Variable(tf.zeros(shape))

def conv_net(x, pool_size, dropout_conv, dropout_fc):    
    # Layer 1: Convolutional. Input = 32x32x1. Output = 28x28x6.
    conv1_W = weigths_variable_conv(kernel_size, 3, conv_output[0], weights_mean, weights_stddev)
    conv1_b = bias_variable(conv_output[0])
    conv1   = conv2d(x, conv1_W, conv1_b)
    # Activation
    conv1 = tf.nn.relu(conv1)
    # Pooling. Input = 28x28x6. Output = 14x14x6.
    conv1 = max_pool(conv1, pool_size, 'VALID')
    # Dropout
    conv1 = tf.nn.dropout(conv1, dropout_conv)
    
    # Layer 2: Convolutional and activation. Output = 10x10x16.
    conv2_W = weigths_variable_conv(kernel_size, conv_output[0], conv_output[1], weights_mean, weights_stddev)
    conv2_b = bias_variable(conv_output[1])
    conv2 = conv2d(conv1, conv2_W, conv2_b)
    # Activation
    conv2 = tf.nn.relu(conv2)
    # Pooling. Input = 10x10x16. Output = 5x5x16.
    conv2 = max_pool(conv2, pool_size, 'VALID')
    # Dropout
    conv2 = tf.nn.dropout(conv2, dropout_conv)


    # Flatten. Input = 5x5x16. Output = 400.
    fc0   = flatten(conv2)
    #fc_len = fc0.get_shape()[1]
    fc_len = 400
    
    # Layer 3: Fully Connected. Input = 400. Output = 120.
    fc1_W = weights_variable_fc(fc_len, fc_output[0], weights_mean, weights_stddev)
    fc1_b = bias_variable(fc_output[0])
    fc1   = tf.matmul(fc0, fc1_W) + fc1_b
    # Activation.
    fc1    = tf.nn.relu(fc1)
    # Dropout
    fc1    = tf.nn.dropout(fc1, dropout_fc)

    # Layer 4: Fully Connected. Input = 120. Output = 84.
    fc2_W  = weights_variable_fc(fc_output[0], fc_output[1], weights_mean, weights_stddev)
    fc2_b  = bias_variable(fc_output[1])
    fc2    = tf.matmul(fc1, fc2_W) + fc2_b
    # Activation.
    fc2    = tf.nn.relu(fc2)
    # Dropout
    fc2    = tf.nn.dropout(fc2, dropout_fc)
    

    # Layer 5: Fully Connected. Input = 84. Output = n_classes.
    fc3_W  = weights_variable_fc(fc_output[1], fc_output[2], weights_mean, weights_stddev)
    fc3_b  = tf.Variable(tf.zeros(fc_output[2]))
    logits = tf.matmul(fc2, fc3_W) + fc3_b
    
    return logits
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the test set but low accuracy on the validation set implies overfitting.

````python
# Training parameters

training_epochs = 50
batch_size = 128
initial_rate = 0.001
rate = initial_rate
len_train = len(X_train)
pool_size = 2

### Train your model here.

### Training pipeline
x = tf.placeholder(tf.float32, (None, 32, 32, 3))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, n_classes, on_value=1., off_value=0., axis=-1)

logits = conv_net(x, pool_size, dropout_conv, dropout_fc)
#cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits, one_hot_y)
loss_operation = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits, one_hot_y))
#loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)
````

````python
### Model evaluation

correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, batch_size):
        batch_x, batch_y = X_data[offset:offset+batch_size], y_data[offset:offset+batch_size]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````

````python
### Train the model
### Calculate and report the accuracy on the training and validation set.
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    
    print("Training...")
    for i in range(training_epochs):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, batch_size):
            end = offset + batch_size
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y})
            
        validation_accuracy = evaluate(X_valid, y_valid)
        print("EPOCH {} ...".format(i+1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print()
        
    saver.save(sess, './lenet')
    print("Model saved")
````
